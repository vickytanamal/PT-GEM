```python
# Import library

import pandas as pd
```

```python
# Load dataset

df1 = pd.read_excel("Case Study Dataset.xlsx", sheet_name='Stock In (Purchase Order)')
df2 = pd.read_excel("Case Study Dataset.xlsx", sheet_name='Stock Out (Used Sparepart)')
```

```python
# Check dimension of data

df1.shape, df2.shape
```

```python
# Check null values Stock in

df1.isnull().sum()
```

```python
# Check null values Stock out

df2.isnull().sum()
```

```python
# Merge dataframes based on different column names

merged_df = pd.merge(df2, df1, left_on='internal_reference', right_on='Product Code', how='left')
```

```python
# Check merged dataset

merged_df.head()
```

```python
# Check missing values

merged_df.isnull().sum()
```

```python
df_missing = merged_df.loc[merged_df['Product Name'].isnull()]
df_missing
```
